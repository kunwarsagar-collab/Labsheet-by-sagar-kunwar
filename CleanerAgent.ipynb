```python
import random

class RoomAgent:
  def __init__(self,room_size=(10,10)):
    self.room_size = room_size
    self.grid=[[random.choice([0,1]) for _ in range(room_size[1]) for _ in range(room_size[0])]]
    self.current_position=(random.randint(0,room_size[0]-1),random.randint(0,room_size[1]-1))

  def display_room(self):
    for row in self.grid:
      for cell in row:
        print(str(cell), end= " ")
      print("\n")

  def perceive(self):
    x,y=self.current_position
    return self.grid[x][y]

  def act(self):
    x, y = self.current_position
    if self.perceive()==1:
      print(f"cell({x}, {y}) is dirty. cleaning...")
      self.grid[x][y]=0
      print(f"cell({x}, {y}) is now clean.")
    else:
      print(f"cell({x}, {y}) is clean. no need to clean.")

  def move(self):
    x, y = self.current_position
    if y < (self.room_size[1]-1):
      self.current_position=(x,y+1)
    elif x < (self.room_size[0]-1):
      self.current_position=(x+1,0)
    else:
      self.current_position=None

  def is_room_clean(self):
    return all(cell==0 for row in self.grid for cell in row)

  def run(self):
    print("Initial Room:")
    self.display_room()
    step = 0
    while not self.is_room_clean():
      print(f"\nStep {step + 1}:")
      self.perceive()
      self.act()
      self.move()
      step+=1
      if self.current_position is None:
        print("Room is full. No more moves possible.")
        self.current_position = (0,0)
    print("Final Room:")
    self.display_room()

agent = RoomAgent()
agent.run()

  # 2. Covert this room cleaning reflex agent to a memory based model based agent
  # 3. Implement successor func for the water jug problem to generate child notes from the given current note using production rules (4. missionary and cannibal)
```
